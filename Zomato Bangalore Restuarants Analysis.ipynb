# Cleaning rate column

```python
df['rate'].unique() # there are some - NEW,nan,-
```

```python
def handle_this_rate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
```

```python
df['rate'] = df['rate'].apply(handle_this_rate)
```

```python
df['rate']
```

```python
df['rate'].isnull().sum()
```

## Filling Null values in rate column with mean

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)  #fillna() is used to fill missing values(nan value) in dataframe or series.
```

```python
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace=True)  #in this data set there are so many null values are present, so dropping the null values in the overall data set.
```

## listed in city and location are similar let's keep on a track and work with one column

```python
df.Cost2plates.unique()
```

## Removing ,(comma) from cost2plates column

```python
def handle_comma(value):
    value = str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
```

```python
df.Cost2plates = df.Cost2plates.apply(handle_comma)
```

```python
df.Cost2plates.unique()
```

```python
df.head()
```

## cleaning rest type column

```python
df.rest_type.value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending=False)
```

```python
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
```

```python
rest_types_lessthan1000
```

## Making rest_type < 1000 in frequency as others

```python
def handle_this_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
```

```python
df['rest_type']=df['rest_type'].apply(handle_this_rest_type)
```

```python
df['rest_type'].value_counts()
```

```python
df.head()
```

## Cleaning Location column

```python
df['location'].value_counts()
```

```python
location=df['location'].value_counts(ascending=False)
```

```python
location_lessthan300 = location[location<300]
```

```python
def handle_this_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
```

```python
df['location']=df['location'].apply(handle_this_location)
```

```python
df['location'].value_counts()
```

```python
df.head()
```

## Cleaning cuisines column

```python
df['cuisines'].value_counts(ascending=False)
```

```python
cuisines= df['cuisines'].value_counts(ascending=False)
cuisines_lessthan100=cuisines[cuisines<100]
```

```python
def handle_this_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'ohers'
    else:
        return value
```

```python
df['cuisines']=df['cuisines'].apply(handle_this_cuisines)
```

```python
df['cuisines'].value_counts()
```
